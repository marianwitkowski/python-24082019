```python
def detect_change(x):
    try:
        return float(x)>=0
    except:
        return np.nan

df = pd.read_csv("f500.csv", 
                 nrows = 30,
                 converters = { "revenue_change" : detect_change },
                 dtype ={ "revenues" : "float64" },
                 index_col='company', 
                 verbose = True,
                 usecols=['company', 'country', 'rank', 'revenues', 
                          'revenue_change', 'profits','profit_change'])
```
