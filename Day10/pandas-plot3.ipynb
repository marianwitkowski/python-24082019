```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://bit.ly/2nBqflH", index_col='country')
df
```

```python
years = df.columns.str.strip('gdpPercap_')
years
```

```python
df.columns = years.astype(int)
df
```

```python
df.loc['Australia'].plot(figsize=(20,10))
```

```python
df.T.plot(figsize=(20,10), grid=True)
plt.ylabel('GDP per capita')
```

```python
plt.style.use('ggplot')
df.T.plot(kind='bar', figsize=(15,15), grid=False)
plt.ylabel('GDP per capita')
```
