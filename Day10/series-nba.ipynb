```python
series = df["Height"]
```

```python
def convert_to_cm(x): # zmiana 6-2 na 6ft i 2in i w koncu na cm
    try:
        list = x.split("-")
        return int(list[0])*30.48 + int(list[1])*2.54 
    except Exception as e:
        return 0

series_cm = series.apply(convert_to_cm)
```
