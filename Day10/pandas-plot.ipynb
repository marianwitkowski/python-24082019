```python
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
data = { 
            "Stopa bezrobocia": [5, 5.5, 4, 3.4, 5],
            "Indeks GPW": [1500, 1400, 1600, 1800, 1200]
       }
df = pd.DataFrame(data, columns=["Stopa bezrobocia","Indeks GPW"])
df.head()
```

```python
df.plot(x='Stopa bezrobocia', y='Indeks GPW', kind='scatter')
```

```python
data = { 
            "Rok": [2010, 2011, 2012, 2013, 2014],
            "Populacja": [38, 38.3, 38.4, 38.2, 38.8]
       }
df = pd.DataFrame(data, columns=["Rok","Populacja"])
df.plot(x='Rok', y='Populacja', kind='line', xticks=[2010, 2011, 2012, 2013, 2014])
```

```python
data = { 
            "Kraj": ['PL', 'DE', 'US', 'FR'],
            "GPD": [35000, 45000, 65432, 33600]
       }
df = pd.DataFrame(data, columns=["Kraj","GPD"])
df.plot(x='Kraj', y='GPD', kind='bar', figsize=(10,7))
```

```python
df = pd.DataFrame([300, 500, 700], columns=["Tasks"], index=['Pending', 'Ongoing', 'Completed'])
df.plot.pie(y='Tasks', figsize=(8,8), startangle=90, autopct="%1.3f%%")
```
